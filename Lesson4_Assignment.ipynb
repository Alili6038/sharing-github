````python
# you may need import some package
import random

x = ['快递太慢了！','不好吃','特别难吃','要齁死我了','很划算','下次还来','味道很不错！','香']
y = ['差评','差评','差评','差评','好评','好评','好评','好评']


    
def shuffle(x,y):
     """
     This function returns two lists with their elements in the same random order
     """
    
    
    
    # your code here
    seed_num = random.randint(0,100)
    random.seed(seed_num)
    random.shuffle(x)
    random.seed(seed_num)
    random.shuffle(y)

    
    
    return x,y

# print result for certify
shuffle(x,y)
for i,j in zip(x,y):
    print(i,':',j)
````
