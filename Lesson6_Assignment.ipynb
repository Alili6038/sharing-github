那为什么在wrapper的函数后面要return wrapper呢

正常情况下定义一个函数以后，里面有打印的话

使用函数的时候不是可以直接用打印来输出结果嘛

```python
from datetime import datetime

def log(func):
    def wrapper(*args, **kw): # 固定写法
        print('calling function:', func.__name__, '\t at ', datetime.now())
        return func(*args, **kw) # ??????????????不懂 ??????????????????????
    return wrapper  # ?????????????????? 

@log
def add_user(name):
    # some code
    print('create user:', name)

@log
def remove_user(name):
    # some code
    print('remove user:', name)
    
add_user('chen'), remove_user('song')
```
