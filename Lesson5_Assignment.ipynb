1. 验证多重继承的属性规则，请选择
- 就近原则
- 随机原则
- 类定义的先后原则

```python
class A0:
    name = 'A0'
    
class A2(A0):  # 继承A0, name = 'A0', num = 2
    num = 2
    
class A1:
    num = 1
    
class A3:
    name = 'A3'
    
# class A6:
#     name = 'A6'
    
class A4(A1,A2,A3): # 继承A1，A2(继承A0), A3
    pass

class A5(A3,A0,A1):
    pass


ins = A4()
ins1 = A5()
print(ins.name) #  类定义的先后原则，先继承A1然后是A2，A2继承A0，所以输出name根据A0中的name来，name = 'A0'
print(ins.num) #  类定义的先后原则，先继承A1，所以输出num根据A1内的num来，num = 1

print(ins1.name) # 同上，先继承A3,name根据A3中的name来，name = 'A3'
print(ins1.num) # 同上，直到继承到A1时，才有num这个attribute,所以输出是A1的num = 1
```
